```python
# PyPI doesn't store a lot of historical data, so we get the last 3 months

pypi_packages = [
    "openfe",
    "gufe",
    "kartograf",
    "cinnabar",
    "pdbinf",
    "exorcist"
]

pypi_dfs = []

for package in pypi_packages:
    df_ = pypistats.overall(
        package, total=True,
       mirrors=True, format="pandas",
       start_date='2022-11-01',
       end_date='2023-10-31'
    )
    df_["Package"] = package
    pypi_dfs.append(df_)

pypidf = pd.concat(pypi_dfs)
```

```python
sorted([x for x in pypidf.date.unique() if isinstance(x, str)])
```

```python
pypidf.groupby("category").get_group("Total")
```

```python
pypi_total = pypidf.groupby("category").get_group("Total").sum().downloads
pypi_total_downloads = int(pypi_total) / 5
print(pypidf)
print(pypi_total_downloads)
```

## conda-forge dependants

```python
import tqdm
import subprocess

dependants = set()
for package in tqdm.tqdm(["openmmtools"]):
    toggle = False
    proc = subprocess.run(
        f"mamba repoquery whoneeds {package} -c conda-forge --quiet",
        shell=True,
        capture_output=True,
        text=True,
    )
    for line in proc.stdout.split("\n"):
        if line.strip().startswith("Name"):
            toggle = True
            continue
        if toggle:
            fields = line.strip().split()
            if len(fields) > 1:
                dependants.add(fields[0])
```

```python
import json

all_public_repos = set()
n_private_dependants = 0
for package in tqdm.tqdm(openff_packages):
    if package.endswith("-base"):
        proc = subprocess.run(
            f'github-dependents-info --repo openforcefield/{package} --sort name --json',
            capture_output=True,
            text=True,
            shell=True,
        )
        contents = json.loads(proc.stdout)
        public_repos = [
            x["name"]
            for x in contents["all_public_dependent_repos"]
        ]
        all_public_repos |= set(public_repos)
        n_private_dependants += contents["private_dependents_number"]
        
```

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from github import Github

%matplotlib inline
```

```python
git = Github(os.environ['GH_API_TOKEN'])
```

```python
org = "choderalab"
```

```python
all_repos = [
    repo.name for repo in git.get_user(org).get_repos()
]
```

```python
all_repos
```

```python
all_unique_counts = 0

for repo_name in all_repos:
    repo = git.get_repo(f"{org}/{repo_name}")
    contents = repo.get_views_traffic(per="week")
    last_fortnight = contents["views"][-2:]
    unique_counts = sum([x.uniques for x in last_fortnight])
    all_unique_counts += unique_counts
```
